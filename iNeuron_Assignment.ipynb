```python
def expand_contractions(text):
    # Expand contractions in the text
    expanded_text = contractions.fix(text)
    return expanded_text
```

Removing StopWord

```python
STOPWORDS = set(stopwords.words("english"))

def remove_STOPWORDS(text):
    return ' '.join([word for word in text.split() if word not in STOPWORDS])

df["No STOPWORDS"] = df["text_lower"].apply(lambda x: remove_STOPWORDS(x))

df["text_lower"]
```

Remove words and digits containing digits

```python
def remove_words_with_digits(text):
    # Use regular expression to remove words and digits containing digits
    text_without_words_with_digits = re.sub(r'\w*\d\w*|\d', '', text)
    return text_without_words_with_digits

# Apply the removal function to the specified column
df["rmv the wrd wth dgt"] = df['No STOPWORDS'].apply(remove_words_with_digits)
df["rmv the wrd wth dgt"]
```

No special char

```python
def remove_spl_char(text):
    text = re.sub('[^a-zA-Z0-9]' , ' ' , text)
    text = re.sub('\s+' , ' ', text)
    return text

df['No special char'] = df["rmv the wrd wth dgt"].apply(lambda x: remove_spl_char(x))
df['No special char']
```

Removal of Frequent words

```python
from collections import Counter
cnt = Counter()
for text in df["No special char"].values:
    for word in text.split():
        cnt[word] += 1
        
cnt.most_common(10)
```

```python
FREQWORDS = set([w for (w, wc) in cnt.most_common(10)])
def remove_freqwords(text):
    """custom function to remove the frequent words"""
    return " ".join([word for word in str(text).split() if word not in FREQWORDS])

df["text_wo_stopfreq"] = df["No special char"].apply(lambda text: remove_freqwords(text))
df["text_wo_stopfreq"]
```

Removing emojies

```python
def remove_emojis(text):
    # Regular expression to match emojis
    emoji_pattern = re.compile("["
        u"\U0001F600-\U0001F64F"  # Emoticons
        u"\U0001F300-\U0001F5FF"  # Symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # Transport & map symbols
        u"\U0001F700-\U0001F77F"  # Alphabetic presentation forms
        u"\U0001F780-\U0001F7FF"  # Geometric shapes
        u"\U0001F800-\U0001F8FF"  # Supplemental Arrows-A
        u"\U0001F900-\U0001F9FF"  # Supplemental Arrows-B
        u"\U0001FA00-\U0001FA6F"  # Supplemental Symbols and Pictographs
        u"\U0001FA70-\U0001FAFF"  # Chess Symbols
        u"\U0001F004-\U0001F0CF"  # Extended emoticons
        u"\U0001F170-\U0001F251"  # Enclosed characters
        "]+", flags=re.UNICODE)
    
    # Remove emojis from the text
    return emoji_pattern.sub(r'', text)

df['removed_emojis'] = df["text_lamm"].apply(remove_emojis)
df['removed_emojis']
```

Sure, let's compare One-Hot Encoding, Bag of Words, N-grams, and TF-IDF in terms of advantages and disadvantages:

### One-Hot Encoding:

#### Advantages:
1. **Simplicity:** One-Hot Encoding is straightforward and easy to understand, making it a good choice for simple tasks.
2. **No Weighting:** Each word has equal importance, making it suitable when word frequency is not a critical factor.
3. **Sparse Representation:** It results in a sparse matrix, which is memory-efficient, especially when dealing with large datasets.
4. **Independence of Vocabulary Size:** It doesn't depend on the vocabulary size, making it easy to handle new words without modifying the encoding.
5. **Interpretability:** The binary representation makes it interpretable, and the absence or presence of a feature is clear.

#### Disadvantages:
1. **Lack of Semantic Information:** One-Hot Encoding doesn't capture the semantic relationships between words.
2. **High Dimensionality:** The encoded vector is very high-dimensional, especially with a large vocabulary, leading to computational inefficiency.
3. **No Contextual Information:** It treats each word in isolation, ignoring the context in which the words appear.
4. **Doesn't Consider Word Frequency:** The model treats all words as equally important, which may not be suitable for tasks where word frequency matters.
5. **Not Suitable for Continuous Data:** It is primarily designed for categorical data and may not be appropriate for continuous data.

### Bag of Words:

#### Advantages:
1. **Simplicity and Efficiency:** Similar to One-Hot Encoding, Bag of Words is simple and computationally efficient.
2. **Flexibility:** It can be applied to various natural language processing tasks, such as sentiment analysis and text classification.
3. **Capture Word Frequency:** It considers the frequency of words in a document, providing more information than One-Hot Encoding.
4. **Sparse Representation:** Like One-Hot Encoding, it results in a sparse representation, saving memory.
5. **Language Agnostic:** It can be applied to texts in different languages without modification.

#### Disadvantages:
1. **Loss of Word Order:** Bag of Words discards the order of words, leading to a loss of important sequential information.
2. **Ignores Semantics:** Similar to One-Hot Encoding, Bag of Words ignores the semantic relationships between words.
3. **Fixed Size Vector:** It generates a fixed-size vector regardless of the document length, potentially losing information from longer documents.
4. **Equal Weighting:** Like One-Hot Encoding, it treats all words as equally important.
5. **Out-of-Vocabulary Words:** It struggles with out-of-vocabulary words as it relies on a predefined vocabulary.

### N-grams:

#### Advantages:
1. **Captures Local Context:** N-grams capture local word relationships, providing some contextual information.
2. **Variable Context Size:** N-grams allow flexibility in capturing different levels of context by adjusting the "N" value.
3. **Preserves Some Order Information:** Unlike Bag of Words, N-grams preserve some order information within a specified context window.
4. **Better for Small Datasets:** In certain cases, N-grams can perform better than Bag of Words with small datasets as they capture local relationships.
5. **Language Agnostic:** Similar to Bag of Words, N-grams can be applied across different languages.

#### Disadvantages:
1. **High Dimensionality:** As with One-Hot Encoding and Bag of Words, N-grams can result in high-dimensional feature spaces.
2. **Fixed Context Size:** The choice of "N" determines the context size, which may not always be optimal for capturing long-range dependencies.
3. **Sparse Representation:** Like the previous methods, N-grams lead to sparse representations, which might be inefficient for large datasets.
4. **Semantic Gaps:** There can still be gaps in capturing semantic relationships between distant words in a document.
5. **Data Sparsity:** In the case of rare n-grams, the data can be sparse, affecting the model's generalization.

### TF-IDF (Term Frequency-Inverse Document Frequency):

#### Advantages:
1. **Considers Word Importance:** TF-IDF considers not only the frequency of words but also their importance in the entire document corpus.
2. **Reduces Common Word Weight:** Common words with high frequency across all documents are downweighted, reducing their impact.
3. **Capture Contextual Importance:** By considering both term frequency and inverse document frequency, TF-IDF captures the contextual importance of words.
4. **Variable Vector Size:** The vector size varies based on the document length, providing a more flexible representation.
5. **Efficient Handling of Stop Words:** Stop words are often downweighted due to their high frequency, mitigating their impact.

#### Disadvantages:
1. **Ignores Word Order:** Similar to Bag of Words and N-grams, TF-IDF ignores the order of words.
2. **Complexity:** The TF-IDF formula involves multiple calculations, making it computationally more expensive than simpler methods.
3. **Language Dependency:** Like Bag of Words and N-grams, TF-IDF can be language-dependent and might require adjustments for different languages.
4. **Loss of Some Local Context:** While capturing global importance, TF-IDF might lose some local context.
5. **Sensitivity to Document Length:** TF-IDF is sensitive to document length, and longer documents might have higher term frequencies.

In summary, the choice between One-Hot Encoding, Bag of Words, N-grams, and TF-IDF depends on the specific requirements and characteristics of the natural language processing task at hand. Each method has its strengths and weaknesses, and the trade-offs should be carefully considered based on the context of the problem being addressed.
